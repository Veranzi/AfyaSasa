```python
# importing packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading dataset
data1=pd.read_excel('/content/Resources Inventory Cost Sheet.xlsx')
data2=pd.read_excel('/content/Treatment Costs Sheet.xlsx')
```

```python
#displyaing first 5 rows for Resource Inventory Cost Sheet
data1.head()
```

```python
#merging the datasets
data= pd.merge(data1, data2, on=['Facility', 'Region'], how='inner')
```
