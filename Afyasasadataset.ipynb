**UNDERSTANDING DATA**

```python
# importing packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading dataset
data=pd.read_excel('/content/Ovarian Cyst Track Data.xlsx')
```

```python
#displays first 5 rows
data.head()
```

```python
#check sum of null entries
data.isnull().sum()
```

```python
data['Date of Exam'] = pd.to_datetime(data['Date of Exam'], format='%Y-%m-%d', errors='coerce')
```

**DATA ANALYSIS AND VISUALISATION**

```python
#datatypes
data.dtypes
```

```python
#displays columns of the dataframe
data.columns
```

```python
#bar plots for cyst size by Region
plt.figure(figsize=(10,3))
sns.barplot(data,x = 'Region', y='Cyst Size cm')
plt.title('Cyst Size by Region')
```

```python
#bar plots for Cyst Size by Menopause Status
plt.figure(figsize=(10,3))
sns.barplot(data,x = 'Menopause Status', y='Cyst Size cm')
plt.title('Cyst Size by Menopause Status')
```

```python
#pie charts showing distribution of ovarian cyst in regions
region_counts = data['Region'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Ovarian Cyst in regions')
plt.show()
```

```python
#pie charts showing recommended management distribution
management_counts = data['Recommended Management'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(management_counts, labels=management_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Recommended Management Distribution')
plt.show()
```

```python
#pie charts showing distribution of Ultrasound Features
features_counts = data['Ultrasound Features'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(features_counts, labels=features_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Ultrasound Features')
plt.show()
```
